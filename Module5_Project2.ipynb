### Upload data set

Run the following lines of code to import Pandas and Matplotlib libraries and upload the data set. 

```python
# import Pandas and Matplotlib libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# upload firm's data
firm_data_df = pd.read_csv('DAT325_HR_Data_Firm.csv')

# report rows and columns in the data set
firm_data_shape = firm_data_df.shape
firm_data_rows, firm_data_cols = firm_data_shape[0], firm_data_shape[1]

print(f'firm data:  rows={firm_data_rows}, cols={firm_data_cols}')
```

## Step 1: Data Quality for Quantitative Variables (Age)

#### 1-1 Missing Values
In this section of data quality checks, you will document evidence of any missing Age values in the data set. Run the code below, review the output (it will print true if there are missing values) and document evidence of missing values.

```python
# this will print true if there are missing values
missing_value_age = firm_data_df['Age'].isnull().any()
print(f'Age has missing values?\n{missing_value_age}')

# this prints rows that have missing values for Age
print(firm_data_df[firm_data_df['Age'].isnull()].to_string())
```

#### 1-2 Incorrect Values
In this step, you will use the describe function in Pandas library to evaluate the quality of variable Age in firm's data. This function prints statistics of Age variable that you can review and validate. Run the step below and identify incorrect minimum value in the variable.

```python
# print stats for variable Age
stats_age = firm_data_df['Age'].describe()
print(f'stats Age \n{stats_age}')

# this prints rows that have negative values for Age
print(firm_data_df[firm_data_df['Age']<0].to_string())
```

#### 1-3 Outlier Values
In this step, you will use the percentile function and a histogram visual to evaluate any positive outliers of variable Age in firm's data. The percentile function is coded to show top 10% values of Age that you can review and validate. You can also review the plot to identify any positive outliers. Run the code below, review the output and document any outliers.

```python
age_top_10_df = firm_data_df['Age'][firm_data_df['Age'] >= firm_data_df['Age'].quantile(0.9)]
age_top_10_df = age_top_10_df.sort_values()
print(age_top_10_df)

plt.hist(firm_data_df['Age'], bins=20)
plt.show() 

# this prints rows that have outlier values for Age, assuming that 110 is selected as outlier threshold
print(firm_data_df[firm_data_df['Age']>110].to_string())
```
